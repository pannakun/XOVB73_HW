Create a Person class with the following attributes:

*   based on the number of children store the number of family members in an attribute  (- this does not have to be included in the init arguments as it is a calculated value)


(4 points)




```python
class Person():
  def __init__(self, name, age, married, n_children, n_cars, money):
    self.name = name
    self.age = age
    self.married = married
    self.n_children = n_children
    self.n_cars = n_cars
    self.money = money

  def family(self):
    if self.married == True:
      family=self.n_children+2
    else:
      family=self.n_children+1
    return family
```

###A) (5 points)
 Create an Animal class and choose 3 animals that have some things in common (characteristics and behaviour)!

Write the Parent class and the derived classes of the species!

The Child classes should inherit some attributes and override some methods of the Parent class, but should also extend the state with own attributes in each subclass and extend the behaviour with own methods.

Decide yourself:

which methods belong to parent


which should be overridden


which attributes are shared. Explain your choice.

Parent Animal class is for mammals on the farm.

The chosen animals: Cow, Horse, Sheep

The shared attributes are: name, species, color, diet and number of legs, since these are attributes of any animal in the word.

The chosen animals sleep while standing, so this method can be shared, for eating it has a shared part, but each animal has it's own diet, so other food can be added to that. The sound is different for each species so it's passed in the parent class.

```python
class Animal():
  def __init__(self, name, species, color, diet = "Herbivore", n_legs = 4):
    self.name = name
    self.species = species
    self.color = color
    self.diet = diet
    self.n_legs = n_legs

  def sleep(self):
    return print(f"{self.name} is sleeping while standing.")

  def eat(self):
    return f"{self.name} is eating grass"

  def sound(self):
    pass

  def __str__(self):
    return f"{self.name} is a {self.color} {self.species}."

class Cow(Animal):
  def __init__(self, name, species, color, diet, n_legs, milk_l_day):
    super().__init__(name, species, color, diet, n_legs)
    self.milk_l_day = milk_l_day

  def eat(self):
    return print(f"{super().eat()} and silage.")

  def sound(self):
    return print(f"{self.name} says Moo!")

  def milk_all_year(self):
    milk_all=365*self.milk_l_day
    return print(f"{self.name} will give {milk_all} liters of milk per year.")


class Horse(Animal):
  def __init__(self, name, species, color, diet, n_legs, rideable, has_shoe):
    super().__init__(name, species, color, diet, n_legs)
    self.rideable = rideable
    self.has_shoe = has_shoe

  def eat(self):
    return print(f"{super().eat()} and hay.")

  def sound(self):
    return print(f"{self.name} says Neigh!")

  def has_owner(self):
    if self.rideable == True or self.has_shoe == True:
      return print(f"{self.name} definitely has an owner.")
    else:
      return print(f"{self.name} is probably a wild horse.")


class Sheep(Animal):
  def __init__(self, name, species, color, diet, n_legs, fur_to_cut):
    super().__init__(name, species, color, diet, n_legs)
    self.fur_to_cut = fur_to_cut

  def eat(self):
    return print(f"{super().eat()} and weeds.")

  def sound(self):
    return print(f"{self.name} says Baa!")

  def trim(self):
    if self.fur_to_cut == True:
      return print(f"It's time for a trim for {self.name}.")
    else:
      return print(f"{self.name} doesn't need a trim.")
```

```python
h1=Horse("Spirit","Mustang horse","buckskin","","",False,False)
c1=Cow("Linda","Highland cow","brown","","",25)
s1=Sheep("Rebecca","Merino sheep","white","","",True)
```

Create classes and use inheritance to calculate the area of different shapes (specifically, circle and square, you can add triangle or anything else if you want).

Requirements:


1.	Create a base class called Shape that has a method named area().

This method should not have any implementation
2. Create at least two subclasses:

Circle — which takes a radius as a parameter

Square — which takes a side as a parameter

write the area-calculating function for both (or for all)

3. Create objects of both Circle and Square, and print their areas.

Extra: Store these objects in a list and use a loop to print the area of each shape.

```python
import math
```

```python
class Shape():
  def area(self):
    pass

  def __str__(self):
    return f"Area is: {self.area():.2f}"

class Circle(Shape):
  def __init__(self, r):
    self.r = r

  def area(self):
    return math.pi * self.r ** 2

class Rectangle(Shape):
  def __init__(self, a, b):
    self.a = a
    self.b = b

  def area(self):
    return self.a * self.b

class Square(Rectangle):
  def __init__(self, a):
    super().__init__(a, a)

class Right_Triangle(Shape):
  def __init__(self, h, b):
    self.h = h
    self.b = b

  def area(self):
    return self.h * self.b / 2
```

```python
sq1=Square(4)
c1=Circle(5)
r1=Rectangle(3,4)
t1=Right_Triangle(2,7)
sq2=Square(6)
r2=Rectangle(5,2)
```

```python
print(sq1)
print(c1)
print(r1)
print(t1)
print(sq2)
print(r2)
```

```python
shape_list = [sq1, c1, r1, t1, sq2, r2]
for shape in shape_list:
  print(f"Area of {shape.__class__.__name__} = {shape.area():.2f}")
```

Count character frequency

```python
text = "banana"
```

Expected output: {'b': 1, 'a': 3, 'n': 2}

```python
def counter(text):
  freq = {}
  for char in text:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1
  return freq
```
